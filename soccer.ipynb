```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
players=pd.read_csv('players_fifa23.csv')
teams=pd.read_csv('teams_fifa23.csv')
```

```python
#top 5 valued and salaried


top5value=players.groupby("Name")["ValueEUR"].sum()
top5value=top5value.sort_values(ascending=False)
top5value=top5value[0:5]
top5value.plot(kind='barh')
```

```python

top5wage=players.groupby("Name")["WageEUR"].sum()
top5wage=top5wage.sort_values(ascending=False)
top5wage=top5wage[0:5]
top5wage.plot(kind='barh')
```
